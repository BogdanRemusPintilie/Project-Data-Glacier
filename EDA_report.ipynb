# Data Preparation

```python
import pandas as pd
import numpy as np
```

## Customer Data

```python
# read dataset
missing_val = ['N/a', 'na', 'NA', np.nan]

col_chose = ['fecha_dato', 'ncodpers', 'pais_residencia', 'sexo', 'age', 'antiguedad', 
             'tiprel_1mes', 'indfall', 'tipodom', 'cod_prov', 'nomprov', 'ind_actividad_cliente', 'renta', 'segmento', 
             'ind_ahor_fin_ult1', 'ind_deco_fin_ult1', 'ind_deme_fin_ult1', 'ind_dela_fin_ult1', 'ind_plan_fin_ult1', 'ind_tjcr_fin_ult1']

df_data = pd.read_csv('data/raw/Train.csv', delimiter = ',', na_values = missing_val, usecols = col_chose)
df_data.info()
```

```python
# check all null data
df_data.isnull().sum()
```

```python
# check and change value
df_data['sexo'].fillna('non-biner', inplace = True)
df_data.isnull().sum()
```

```python
# check and change value
df_data['tiprel_1mes'].fillna('R', inplace = True)
df_data.isnull().sum()
```

```python
# check and change value
df_data['segmento'].fillna('other', inplace = True)
df_data.isnull().sum()
```

```python
# filter gross income household
sample = df_data.loc[df_data['renta'].notnull()]
sample.isnull().sum()
```

```python
# filter latin america code province
df_filter = sample.loc[sample['cod_prov'].notnull()]
df_filter.isnull().sum()
```

```python
df_filter.info()
```

```python
# rename column header
col_rename = {'fecha_dato':'date', 'ncodpers':'customer_code', 'pais_residencia':'residence_country_code', 'sexo':'gender', 'age':'age', 'antiguedad':'seniority_in_month', 
              'tiprel_1mes':'customer_type', 'indfall':'dead_status', 'tipodom':'address', 'cod_prov':'code_prov', 'nomprov':'province', 'ind_actividad_cliente':'customer_activity', 
              'renta':'gross_income_usd', 'segmento':'segmentation', 'ind_ahor_fin_ult1':'saving_account', 'ind_deco_fin_ult1':'short_term_deposits', 'ind_deme_fin_ult1':'medium_term_deposits', 
              'ind_dela_fin_ult1':'long_term_deposits', 'ind_plan_fin_ult1':'pensions', 'ind_tjcr_fin_ult1':'credit_card'}
df_rename = df_filter.rename(columns = col_rename)
df_rename.head()
```

```python
# apply lower or upper case in atribute province
df_data = df_rename.copy()
df_data['province'] = df_data['province'].str.lower()
df_data.head()
```

```python
# remove special char in atribute province
df_data['province'] = df_data['province'].str.replace(r'\W', '')
df_data.head()
```

```python
# rename atribute in segmentation
df_data.loc[df_data['segmentation'] == '01 - TOP', 'segmentation'] = 'vip'
df_data.loc[df_data['segmentation'] == '02 - PARTICULARES', 'segmentation'] = 'individual'
df_data.loc[df_data['segmentation'] == '03 - UNIVERSITARIO', 'segmentation'] = 'college_graduated'

df_data.head()
```

```python
# remove special char in atribute province
df_data['segmentation'] = df_data['segmentation'].str.replace(r'\W', '')
df_data.head()
```

```python
# rename atribute in gender
df_data.loc[df_data['gender'] == 'H', 'gender'] = 'male'
df_data.loc[df_data['gender'] == 'V', 'gender'] = 'female'

df_data.head()
```

```python
# rename atribute in customer_type
df_data.loc[df_data['customer_type'] == 'A', 'customer_type'] = 'active'
df_data.loc[df_data['customer_type'] == 'I', 'customer_type'] = 'unactive'
df_data.loc[df_data['customer_type'] == 'P', 'customer_type'] = 'former'
df_data.loc[df_data['customer_type'] == 'R', 'customer_type'] = 'potential'

df_data.head()
```

```python
# rename atribute in gender
df_data.loc[df_data['dead_status'] == 'N', 'dead_status'] = 'alive'
df_data.loc[df_data['dead_status'] == 'V', 'dead_status'] = 'dead'

df_data.head()
```

```python
df_data.info()
```

```python
# change data type

# date convert
df_data['date'] = pd.to_datetime(df_data['date'])

# data convert
data_convert = {'age': int, 'seniority_in_month': int, 'address': int, 'code_prov': int, 'customer_activity': int}

df_data = df_data.astype(data_convert)
df_data.info()
```

```python
# drop duplicates
df_data.drop_duplicates(subset = ['customer_code'])
df_data.info()
```

## Data Manipulations

```python
df_data.describe().round(0)
```

```python
df_man = df_data.loc[df_data['seniority_in_month'] >= 0]
df_man.head()
```

```python
df_man.describe().round(0)
```

```python
# add year category
df_man['year'] = df_man['date'].dt.year
df_man.head()
```

```python
# add month category
df_man['month'] = df_man['date'].dt.month
df_man.head()
```

```python
# add week category
df_man['day_cat'] = df_man['date'].dt.day_of_week
df_man.head()
```

```python
# add day
df_man.loc[df_man['day_cat'] == 0, 'day'] = 'monday'
df_man.loc[df_man['day_cat'] == 1, 'day'] = 'tuesday'
df_man.loc[df_man['day_cat'] == 2, 'day'] = 'wednesday'
df_man.loc[df_man['day_cat'] == 3, 'day'] = 'thursday'
df_man.loc[df_man['day_cat'] == 4, 'day'] = 'friday'
df_man.loc[df_man['day_cat'] == 5, 'day'] = 'saturday'
df_man.loc[df_man['day_cat'] == 6, 'day'] = 'sunday'

df_man.head()
```

```python
# add weekday category
df_man.loc[df_man['day'] == 'monday', 'week_cat'] = 'weekday'
df_man.loc[df_man['day'] == 'tuesday', 'week_cat'] = 'weekday'
df_man.loc[df_man['day'] == 'wednesday', 'week_cat'] = 'weekday'
df_man.loc[df_man['day'] == 'thursday', 'week_cat'] = 'weekday'
df_man.loc[df_man['day'] == 'friday', 'week_cat'] = 'weekday'
df_man.loc[df_man['day'] == 'saturday', 'week_cat'] = 'weekend'
df_man.loc[df_man['day'] == 'sunday', 'week_cat'] = 'weekend'

df_man.head()
```

```python
# add age category
df_man.loc[df_man['age'] <= 25, 'age_category'] = 'young adult (< 26)'
df_man.loc[(df_man['age'] > 25) & (df_man['age'] <= 40), 'age_category'] = 'adult (26 ~ 40)'
df_man.loc[(df_man['age'] > 40) & (df_man['age'] <= 60), 'age_category'] = 'late adult (41 ~ 60)'
df_man.loc[df_man['age'] > 60, 'age_category'] = 'elder (> 60)'

df_man.head()
```

```python
# add gross income category
df_man.loc[df_man['gross_income_usd'] < 68711, 'gross_category_usd'] = 'bellow_average'
df_man.loc[(df_man['gross_income_usd'] >= 68711) & (df_man['gross_income_usd'] < 155952), 'gross_category_usd'] = 'average'
df_man.loc[df_man['gross_income_usd'] >= 155952, 'gross_category_usd'] = 'above_average'

df_man.head()
```

```python
# add seniority cat
df_man.loc[df_man['seniority_in_month'] < 26, 'cus_seniority_cat'] = 'entry_level'
df_man.loc[(df_man['seniority_in_month'] >= 26) & (df_man['gross_income_usd'] <= 140), 'cus_seniority_cat'] = 'mid_level'
df_man.loc[df_man['seniority_in_month'] > 140, 'cus_seniority_cat'] = 'senior_level'

df_man.head()
```

```python
# product filtering
product_filter = ((df_man['saving_account'] > 0) | (df_man['short_term_deposits'] > 0) | (df_man['medium_term_deposits'] > 0) | 
                  (df_man['long_term_deposits'] > 0) | (df_man['pensions'] > 0) | (df_man['credit_card'] > 0))
minimum = df_man.loc[product_filter]

print('before filter: ', df_man.shape)
print('after filter: ', minimum.shape)
```

```python
# save data preparations
fin_dt = minimum.copy()
fin_dt.to_csv('data/processed/data.csv', index = False)
```
